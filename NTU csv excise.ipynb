```python
import csv
dataset = [ ]
with open("tw_population_density.csv","r") as fIn:
    csvIn = csv.reader(fIn)
    for line in csvIn:
        dataset.append(line)
    
```

```python
dataset
```

```python
head=dataset[0]
```

```python
body=dataset[1:]
```

```python
def selectCol(item):
    #print("item,item,int(item[-1])")
    return int(item[-1])
result=sorted(body,key=selectCol,reverse=True)
```

```python
result.insert(0,head)
```

```python
with open("tw_population_density.csv","w") as fOut:
    csvOut = csv.writer(fOut)
    csvOut.writerows(result)
```

```python
# mport requests
# from pyquery import PyQuery as pq

# response = requests.get('https://www.webscraper.io/test-sites/e-commerce/allinone')

# doc = pq(response.text)

# doc('body > div.container > div > div.col-md-9 > div.jumbotron > h1').text() #要變成字串 #從html標籤跳到目標標籤上

# doc('body > div.container > div > div.col-md-9 > div.jumbotron > p').text() #在瀏覽器上 對著標籤 複製選擇器

# doc('body > div.container > div > div.col-md-9 > div.row > div:nth-child(1) > div > div.caption > h4:nth-child(2) > a').text()

# doc('body > div.container > div > div.col-md-9 > div.row > div > div > div.caption > h4:nth-child(2) > a').text() 
# #把":nth-child(1)"刪掉
# #a標籤 代表的是連結

# doc('body > div.container > div > div.col-md-9 > div.row > div > div > div.caption > h4:nth-child(2) > a')

# for eachDoc in doc('body > div.container > div > div.col-md-9 >\
# div.row > div > div > div.caption > h4:nth-child(2) > a').items(): #要對裡面的所有物件迭代
#     print(eachDoc.text())

import requests
from pyquery import PyQuery as pq
response = requests.get('https://www.webscraper.io/test-sites/e-commerce/allinone')
doc = pq(response.text)

for eachMainDoc in doc("#side-menu > li:nth-child(n+2) > a").items():
    print(eachMainDoc.text())

dataset = []
doc.make_links_absolute(base_url = response.url)
for eachMainDoc in doc("#side-menu > li:nth-child(n+2) > a").items():
    mainCateDoc = pq(eachMainDoc.attr("href"))
    mainCateDoc.make_links_absolute()
    for eachSubDoc in mainCateDoc("#side-menu .subcategory-link").items():
        print(eachSubDoc.attr("href"))
        subCateDoc = pq(eachSubDoc.attr("href"))
        # lvl2---------
        #print(len(subCateDoc(".thumbnail")), subCateDoc(".thumbnail").text())
        for eachitemDoc in subCateDoc(".thumbnail").items():
            itemDict = {}
            itemDict["title"] = eachitemDoc(".title").text().strip(".") #前後將不要的東西去掉(變形：lstrip, rstrip)
            itemDict["price"] = float(eachitemDoc(".price").text().replace("$","")) #新增一個Key
            itemDict["stars"] = len(eachitemDoc(".ratings .glyphicon-star"))
            dataset.append(itemDict)
            #print(len(eachitemDoc(".ratings .glyphicon-star"))) #子孫選擇器 #不是要內容而是要算選到幾個，所以不用.text()

```

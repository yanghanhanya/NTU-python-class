```python
import urllib.request
import pandas as pd
```

```python
title = []
date = []
url = []
author = []
```

```python
p = 0  #page
while p != 20: #網頁爬到第20頁
    p = p + 1
    cate = 'cutting-edge/ai/page/' #網址的前面
    page = urllib.request.urlopen("https://technews.tw/category/%s%d"%(cate, p))  #cate帶入％S為一個字符串,p帶入％d為一個整數
    text = page.read().decode("utf8")
    titleurlFind = text.split("bookmark\">")  #切割到哪
    authorFind = text.split("author\">")
    dateFind = text.split("發布日期</span>")
```

```python
n = 0
for line in dateFind[1:]:
    n = n + 1
    splittitle = titleurlFind[n].index("</a>") #"</a>"在第幾個位置，為一個整數
    splitdate = dateFind[n].index("</span>")
    splitauth = authorFind[n].index("</a>")
    spliturl1 = titleurlFind[n-1].index("class=\"entry-title\"><a href=\"")
    spliturl2 = titleurlFind[n-1].index("\" title=\"")
    title.append(titleurlFind[n][:splittitle])
    date.append(dateFind[n][28:splitdate]) #在第28個字後開始是日期
    author.append(authorFind[n][:splitauth])
    url.append(titleurlFind[n-1][spliturl1+29:titleurlFind[n-1].find(title[n-1])-9])
```

```python
AI_dict = {
        "1Title":title,
        "2Author":author,
        "3Date":date,
        "4Link":url
        }
AI_df = pd.DataFrame(AI_dict)
```
